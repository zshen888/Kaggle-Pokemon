```python
# Change and set the index to 'Name'
df = df.set_index('Name')
df.head(n=10)
```

```python
# Remove repeated text in Mega pokemon names
df.index = df.index.str.replace('.*(?=Mega)', '')
df.head(n=10)
```

```python
# Drop col with axis = 1; axis = 0 is for rows
df = df.drop(['#'], axis = 1)
#fill NaN values in Type2 with corresponding values of Type
df['Type 2'].fillna(df['Type 1'], inplace=True)

print('The Cols of the data are: \n', df.columns)
print('The shape of the data is: \n', df.shape)
```
