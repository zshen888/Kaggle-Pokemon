```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import os
print(os.getcwd())
df = pd.read_csv('../Data/Pokemon.csv')
```

```python
print(df.head(n=10))
print(df.tail(n=10))
```

```python
df = df.drop(['#', 'Total', 'Generation', 'Legendary'], 1)
```

```python
sns.jointplot(x='HP', y='Attack', data=df)
plt.show()
```

```python
# Distribution of all numeric stats
sns.boxplot(data=df)
plt.show()
```

```python
print(len(df))
df = pd.melt(df, id_vars=['Name', 'Type 1', 'Type 2'], var_name='Stat')
print(len(df))
# 4800 rows up from 800
```

```python
sns.swarmplot(x='Stat', y='value', data=df, hue='Type 1')
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.ylim(0, 300)
sns.swarmplot(x='Stat', y='value', data= df, hue='Type 1', dodge=True, size=7)
plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.show()
```

```python
# Custom 
sns.set_style("whitegrid")
with sns.color_palette([
    "#8ED752", "#F95643", "#53AFFE", "#C3D221", "#BBBDAF",
    "#AD5CA2", "#F8E64E", "#F0CA42", "#F9AEFE", "#A35449",
    "#FB61B4", "#CDBD72", "#7673DA", "#66EBFF", "#8B76FF",
    "#8E6856", "#C3C1D7", "#75A4F9"], n_colors=18, desat=.9):
    plt.figure(figsize=(12,10))
    plt.ylim(0, 300)
    sns.swarmplot(x='Stat', y='value', data= df, hue='Type 1', dodge=True, size=7)
    plt.legend(bbox_to_anchor=(1,1), loc=2)
    plt.show()
```
